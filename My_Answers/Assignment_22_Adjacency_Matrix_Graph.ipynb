```python
class Graph:
    def __init__(self, vertices):
        self.vertices = vertices
        self.adj_matrix = [[0] * vertices for _ in range(vertices)]

    def add_edge(self, vertex1, vertex2,weight):
        # Add an edge between vertex1 and vertex2
        self.adj_matrix[vertex1][vertex2] = weight
        self.adj_matrix[vertex2][vertex1] = weight

    def remove_edge(self, vertex1, vertex2):
        # Remove the edge between vertex1 and vertex2
        self.adj_matrix[vertex1][vertex2] = 0
        self.adj_matrix[vertex2][vertex1] = 0

    def has_edge(self, vertex1, vertex2):
        # Check if there is an edge between vertex1 and vertex2
        return self.adj_matrix[vertex1][vertex2] >= 1

    def display(self):
        # Display the adjacency matrix
        for row in self.adj_matrix:
            print(" ".join(map(str, row)))

# Example usage:
# Create a graph with 4 vertices
my_graph = Graph(4)

# Add edges
my_graph.add_edge(0, 1,43)
my_graph.add_edge(0, 2,32)
my_graph.add_edge(1, 3,64)
print(my_graph.has_edge(1,0))

# Display the adjacency matrix
my_graph.display()
```
