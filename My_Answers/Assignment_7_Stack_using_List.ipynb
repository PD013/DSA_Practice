```python
# This is Sirs #U can refer ur own code in the bottom
class Stack:
    def __init__(self):
        self.items = []

    def is_empty(self):
        return len(self.items)==0 #Here if the condition is true then it returns True or else False

    def push(self,data):
        self.items.append(data)  #here the item gets inserted at the end using append [Its a list method]

    def pop(self):
        if not self.is_empty():
            return self.items.pop()  #here the item is removed froom the end using pop [Its a list method]
        else:
            raise IndexError("Stack is Empty")
    def peek(self):
        if not self.is_empty():
            return self.items[-1] #here it just shows the last item(which is top of the stack)
        else:
            raise IndexError("Stack is Empty")

    def size(self):
        return len(self.items)
s1=Stack()
s1.push(11)
s1.push(244)
s1.push(30)
print("Top element is",s1.peek())
print("Removed element is",s1.pop())
print("Top element is",s1.peek())
print()
```
