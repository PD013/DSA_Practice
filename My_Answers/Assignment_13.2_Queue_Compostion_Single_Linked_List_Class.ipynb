```python
from Assign3_ANSWER_SIR import *
class Queue:
    def __init__(self):
        self.mylist=SLL()
        self.item_count=0
        # self.front = None # No need because it can be acheived by using start of sll
        self.rear = None
    def is_empty(self):
        return self.mylist.is_empty()

    def enqueue(self,data):
        self.mylist.insert_at_last(data)
        self.item_count +=1
        self.rear = data

    def dequeue(self):
        if not self.is_empty():
            if self.get_front() == self.rear:
                self.rear = None
            self.mylist.delete_first()
            self.item_count -=1

    def get_front(self):
        return self.mylist.start.item

    def get_rear(self):
        return self.rear

    def size(self):
        return self.item_count
q1 = Queue()

q1.enqueue(1)
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

q1.enqueue(2)
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

q1.enqueue(3)
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

q1.enqueue(4)
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

q1.dequeue()
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

print("This is the Size ",q1.size())
```
