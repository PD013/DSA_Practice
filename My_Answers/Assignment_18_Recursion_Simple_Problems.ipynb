```python
def print_numbers(n):
    # Base case: stop when n becomes 0
    if n <= 0:
        return
    else:
        # Recursive case: print the current number and call the function with n-1
        print(n)
        print_numbers(n - 1)

# Example: Print the first 5 numbers
print_numbers(5)
```

```python
def print_odd_num(n): # Here all the odd numbers before the given number would be printed
    if n<=0:
        return
    else:
        if n % 2 == 0:
            print_odd_num(n-1)
        else:
           print_odd_num(n-2)
           print(n)

print_odd_num(13)
```

```python
def print_first_n_odd_numbers(n, current=1): # Here Odd numbers till the given N number will be printed
    if n <= 0:
        return
    else:
        print(current)
        print_first_n_odd_numbers(n - 1, current + 2)

# Example: Print the first 5 odd numbers
print_first_n_odd_numbers(5)
```
