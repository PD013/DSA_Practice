```python
class Queue(list):
    def __init__(self):
        self.counter = 0 # Noo need we can just use len(self.my.list) Only required when using linkedList

    def is_empty(self):
        return not self # This is also a very good way to check

    def enqueue(self,data):
        self.append(data)
        self.counter +=1

    def dequeue(self):
        if not self.is_empty():
            self.pop(0)
            self.counter -=1
        else:
            print("Queue is empty. Cannot dequeue.")

    def get_front(self):
        if not self.is_empty():
            return self[0]
        else:
            raise IndexError("No elemnt in the list")

    def get_rear(self):
        if not self.is_empty():
            return self[-1]
        else:
            raise IndexError("No elemnt in the list")

    def size(self):
        return self.counter # also can use len(self.my.list)

q1 = Queue()

q1.enqueue(1)
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

q1.enqueue(2)
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

q1.enqueue(3)
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

q1.enqueue(4)
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

q1.dequeue()
print("This is the current front element",q1.get_front())
print("This is the current rear element",q1.get_rear())
print()
print()

print("This is the Size ",q1.size())


# The expression `not self.mylist` works by evaluating the truthiness of the list `self.mylist`.
# In Python, an empty list is considered falsy in a boolean context, meaning it evaluates to `False`.
# Therefore, `not self.mylist` is `True` when the list is empty and `False` when it contains elements.
```
