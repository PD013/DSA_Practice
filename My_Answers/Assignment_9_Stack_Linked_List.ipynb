```python
# Sirs code with my snippets
class Node:
    def __init__(self,item=None,next=None):
        self.item=item
        self.next=next
class Stack:
    def __init__(self):
        self.start=None
        self.item_count=0
    def is_empty(self):
        return self.start==None
# Here think of LinkedLis here we are adding new data in form of node at the 1st position
    def push(self,data):
        n=Node(data,self.start)
        self.start=n
        self.item_count+=1
# Here poping is also done for the 1st element as we have doing it in LinkedLis
    def pop(self):
        if not self.is_empty():
            data=self.start.item
            self.start=self.start.next #changing the start to the 2nd node
            self.item_count-=1
            return data
        else:
            raise IndexError("Stack is empty")
    def peek(self): # here the 1st node item is printed
        if not self.is_empty():
            return self.start.item
        else:
            raise self.is_empty()
    def size(self): # For simplifiying counting we had a variable intialized at the start so it becomes easy
        return self.item_count
s1=Stack()
s1.push(10)
s1.push(20)
s1.push(30)
print("Total elements in the stack=",s1.size())
print("Top element on the stack is",s1.peek())
print("Poped element is",s1.pop())
print("Total elements in the stack=",s1.size())
print("Top element on the stack is",s1.peek())
```
