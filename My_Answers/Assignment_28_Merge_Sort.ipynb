```python
def merge_sort(list1):
    if len(list1)>1:
        mid=len(list1)//2
        leftlist=list1[:mid] # Here we are making 2 parts
        rightlist=list1[mid:]

        merge_sort(leftlist) # here recurssion is happening
        merge_sort(rightlist)

        i=j=k=0
        while i<len(leftlist) and j<len(rightlist): # This case will work till there are equal number oof elements on both side
            if leftlist[i]<rightlist[j]: # Compare and add in the original list one by one
                list1[k]=leftlist[i]
                i+=1
            else:
                list1[k]=rightlist[j]
                j+=1
            k+=1
        while i<len(leftlist): # if one side is empty and other side then add whole other side one by one without comparing
            list1[k]=leftlist[i] # This is if the left list is remaining
            i+=1
            k+=1
        while j<len(rightlist):  # This is if the right list is remaining
            list1[k]=rightlist[j]
            j+=1
            k+=1

mylist=[75,29,83,42,16,90,56,34,20,71,88,92,7]
merge_sort(mylist)
print(mylist)
```

```python
# This is my Code
def merge_sort(lst):
    if len(lst)<=1:
        return lst
    else:
        highest = len(lst)-1
        mid = (0+highest//2)
        leshalf = lst[0:mid]
        bighalf = lst[mid:]
        return merge_sort(leshalf)+merge_sort(bighalf)

mylist=[53,11,72,68,41,25,18,37,44,80]
mylist=merge_sort(mylist)
print(mylist)
```
