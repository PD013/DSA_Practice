```python
class Node:
    def __init__(self,prev=None,item=None,next=None):
        self.prev = prev
        self.next = next
        self.item = item

class Deque:
    def __init__(self):
        self.counter = 0
        self.front = None
        self.rear = None

    def is_empty(self):
        return self.counter ==0

    def insert_front(self,data):
        n = Node(None,data,self.front)
        if not self.is_empty():
            self.front.prev = n
        else:
            self.rear = n
        self.front = n
        self.counter +=1

    def insert_rear(self,data):
        n = Node(self.rear,data,None)
        if not self.is_empty():
            self.rear.next = n
        else:
            self.front = n
        self.rear = n
        self.counter +=1

    def delete_front(self):
        if not self.is_empty():
              if self.front == self.rear:
                  self.front = None
                  self.rear = None
              else:
                  self.front = self.front.next
                  self.front.prev = None
              self.counter -=1
        else:
            raise IndexError("The list is empty")

    def delete_rear(self):
        if not self.is_empty():
              if self.front == self.rear:
                  self.front = None
                  self.rear = None
              else:
                  self.rear = self.rear.prev
                  self.rear.next = None
              self.counter -=1
        else:
            raise IndexError("The list is empty")

    def get_front(self):
        if not self.is_empty():
            return self.front.item
        else:
            return None
    def get_rear(self):
        if not self.is_empty():
            return self.rear.item
        else:
            return None

    def size(self):
        return self.counter

    def test(self):
        return self.front.next.item

q2 = Deque()

print(f"Initial size: {q2.size()}\n")

    # Inserting elements
print("Inserting elements:")
q2.insert_front(1)
print(f"Inserted front: 1, Current front: {q2.get_front()}, Current rear: {q2.get_rear()}, Size: {q2.size()}")
q2.insert_rear(2)
print(f"Inserted rear: 2, Current front: {q2.get_front()}, Current rear: {q2.get_rear()}, Size: {q2.size()}")
q2.insert_rear(3)
print(f"Inserted rear: 3, Current front: {q2.get_front()}, Current rear: {q2.get_rear()}, Size: {q2.size()}\n")

# Deleting elements
print("Deleting elements:")
q2.delete_front()
print(f"Deleted front, Current front: {q2.get_front()}, Current rear: {q2.get_rear()}, Size: {q2.size()}")
q2.delete_rear()
print(f"Deleted rear, Current front: {q2.get_front()}, Current rear: {q2.get_rear()}, Size: {q2.size()}\n")

# Inserting more elements
print("Inserting more elements:")
q2.insert_front(4)
print(f"Inserted front: 4, Current front: {q2.get_front()}, Current rear: {q2.get_rear()}, Size: {q2.size()}")
q2.insert_rear(5)
print(f"Inserted rear: 5, Current front: {q2.get_front()}, Current rear: {q2.get_rear()}, Size: {q2.size()}\n")

# Final state
print("Final state:")
print(f"Current front: {q2.get_front()}, Current rear: {q2.get_rear()}, Size: {q2.size()} , Middle element : {q2.test()}")




```
