## 1. **Write a Python script to sort the values in an array.**

> Arrays in Pythoon has to be imported

```python
from array import *

a1 = array('i', [2, 3, 1, 4, 5, 6, 8, 7, 9])
print("THis is the Array which we have to sort :- ",a1)

# Convert the array to a list
arr_to_lst = a1.tolist()
print("This is the list converted from Array:", arr_to_lst)

# Sort the list
sort_lst = sorted(arr_to_lst)
print("This is the sorted list:", sort_lst)

# You can convert the sorted list back to an array
list_to_arr = array('i', sort_lst)
print("This is the array converted back from the sorted list:", list_to_arr)

# To sort the arr its required to convert it to list compulsoryly
```

## 2. **Write a Python script to remove all non-integer values from a list of heterogeneous elements.**


```python
# Input list with heterogeneous elements
input_list = [1, 'apple', 3.14, 'banana', 42, 'cherry', '5']

# Create a new list containing only integers from the input list
integers_only = [x for x in input_list if isinstance(x, int)]

# Print the list containing only integers
print(integers_only)
```

## 3. **Write a Python script to calculate the average of elements in a list.**



```python
def calculate_average(input_list):
    # Check if the list is not empty
    if len(input_list) > 0:
        # Calculate the sum of all elements in the list
        total = sum(input_list)
        # Calculate the average by dividing the total by the number of elements
        average = total / len(input_list)
        return average
    else:
        return 0  # To handle the case of an empty list

# Example usage:
my_list = [1, 2, 3, 4, 5, 9, 10, 3]
result = calculate_average(my_list)
print("Average:", result)
```

## 4. **Write a Python script to create a list of the first N prime numbers.**


```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

def generate_primes(N):
    prime_list = []
    num = 2
    while len(prime_list) < N:
        if is_prime(num):
            prime_list.append(num)
        num += 1
    return prime_list

# Example usage:
N = 10  # Change N to the desired number of prime numbers
prime_numbers = generate_primes(N)
print("First", N, "prime numbers:", prime_numbers)
```

```python
def isprime(num):
    """
    Check if a number is prime.
    A number is not prime if remainder(%) is zero when divided by any num (that num cannot be 1 or itself)
    Parameters:
    num (int): The number to check for primality.

    Returns:
    bool: True if the number is prime, False otherwise.
    """
    for i in range(2, num):
        if num % i == 0:
            return False
    return True

def primeGenerator(n):
    """
    Generate the first 'n' prime numbers using a generator.
    Here We used Yeild which ends the program at the yeild statement but starts from there only when called
    again it becomes generator/iterator on which we have to iterate to find all the values
    Parameters:
    n (int): The number of prime numbers to generate.

    Yields:
    int: The next prime number in the sequence.

    Returns:
    None
    """
    num = 2
    while n:
        if isprime(num):
            yield num
            n -= 1
        num += 1

# Example usage:
it = primeGenerator(10)
print("First 10 prime numbers:", end=' ')
for e in it:
    print(e, end=' ')
```

## 5. **Write a Python script to create a list of the first N terms of a Fibonacci series.**


```python
def generate_fibonacci(N):
    fibonacci_list = []
    a, b = 0, 1
    for i in range(N):
        fibonacci_list.append(a)
        a, b = b, a + b      # Lets take N =5 and do soome itterations
#       0, 1 = 1 , 0 +1    //1st Iteration done fibonacci_list = [0]
#            a =1 , b = 1  == New a and b
#       1, 1 = 1 , 1 + 1    //2nt Iteration done fibonacci_list = [0,1]
#               Hence B becomes the new A and A + B becomes the new B

    return fibonacci_list

# Example usage:
N = 10  # Change N to the desired number of Fibonacci terms
fibonacci_series = generate_fibonacci(N)
print("First", N, "Fibonacci terms:", fibonacci_series)
```
