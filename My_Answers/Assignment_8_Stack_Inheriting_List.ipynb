```python
class List:

    def __init__(self):
        self._items = []

class Stack(List):

    def is_empty(self):
        return len(self._items)==0 # here if 0 it acts as a False any num will act as True

    def push(self,data):
        self._items.append(data)

    def pop(self):
        if not self.is_empty:
            return self._items.pop()
        else:
            raise IndexError("Stack is empty")

    def peek(self):
        if not self.is_empty:
            return self._item[-1]
        else:
            raise IndexError("Stack is empty")

    def size(self):
        return len(self._items)

s1=Stack()
s1.push(10)
s1.push(20)
s1.push(30)
print("top value=",s1.peek())
print()
```
