```python
# Sirs Code but here I have written all the code without any help it felt easy compared to previous Assignment
def sumN(n):
    if n==1:
        return 1
    return n + sumN(n-1)

def sumNOdd(n):
    if n==1:
        return 1
    return 2*n-1 + sumNOdd(n-1)

def sumNEven(n):
    if n==1:
        return 2
    return 2*n + sumNEven(n-1)

def fact(n):
    if n==0:
        return 1
    return n * fact(n-1)

def sumNSquare(n):
    if n==1:
        return 1
    return n*n + sumNSquare(n-1)
print("Sum is",sumNSquare(5))
```

```python
def sum_n_number(n):
    if n == 1:
        return 1
    else:
        s = n + sum_n_number(n-1)
        return s

p = sum_n_number(4)
print(p)
```
