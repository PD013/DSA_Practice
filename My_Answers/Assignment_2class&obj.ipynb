```python
class Circle:
    def __init__(self, radius):
        self._radius = radius  # Use an underscore prefix to indicate it's a protected variable

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, radius):
        if radius >= 0:
            self._radius = radius
        else:
            raise ValueError("Radius must be non-negative")

    def getArea(self):
        area = 3.14 * (self._radius ** 2)
        return area

    def getCircumference(self):
        circumference = 2 * 3.14 * self._radius
        return circumference

A1 = Circle(4)
A1.radius
A1.getArea()
A1.getCircumference()

```

```python
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    @property
    def dimensions(self):
        return self.length, self.breadth

    @dimensions.setter
    def dimensions(self, values):
        length, breadth = values
        if length > breadth:
            self.length = length
            self.breadth = breadth
        else:
            raise ValueError("Length must be greater than the breadth")

    def showDimensions(self):
        print(f"The Dimensions of the Rectangle are {self.length} Length and {self.breadth} Breadth")

    def getArea(self):
        area = self.length * self.breadth
        return area

A1 = Rectangle(5, 3)
A1.showDimensions()  # Display dimensions
A1.dimensions = (6, 4)  # Set new dimensions
A1.showDimensions()  # Display updated dimensions
area = A1.getArea()
print(f"The area of the Rectangle is: {area}")
```

```python
# Define a Team class
class Team:
    def __init__(self):
        # Initialize an empty list to store team member names
        self.members = []

    def input_members(self):
        # Ask the user for the number of team members they want to add
        num_members = int(input("Enter the number of team members: "))

        # Use a loop to input each team member's name
        for i in range(num_members):
            member_name = input(f"Enter the name of team member {i+1}: ")

            # Add the member's name to the list of team members
            self.members.append(member_name)

    def display_members(self):
        # Check if there are team members in the list
        if self.members:
            print("Team Members:")

            # Loop through the list and display each member's name
            for member in self.members:
                print(member)
        else:
            # If the list is empty, indicate that there are no team members
            print("No team members in the list.")

# Create an instance of the Team class
team1 = Team()

# Input team member names
team1.input_members()

# Display team member names
team1.display_members()
```

### 2. Creating a subclass (Inheritance)

```python
# Define a parent class named 'Animal'
class Animal:
    # Constructor for Animal class, takes a 'name' parameter
    def __init__(self, name):
        self.name = name  # Assign the 'name' parameter to the 'name' attribute

    # Method to make the animal speak
    def speak(self):
        return f"{self.name} says hello!"  # Return a greeting with the 'name'

# Define a child class named 'Dog' that inherits from 'Animal'
class Dog(Animal):
    # Override the 'speak' method from the parent class
    def speak(self):
        return f"{self.name} barks!"  # Return a message indicating barking

# Create an instance of the 'Dog' class with the name "Charlie"
dog = Dog("Charlie")

# Call the 'speak' method of the 'dog' object
print(dog.speak())  # Output will be "Charlie barks!"
```

### 3. Using the super() function

```python
# Define a parent class named 'Animal'
class Animal:
    # Constructor for Animal class, takes a 'name' parameter
    def __init__(self, name):
        self.name = name  # Assign the 'name' parameter to the 'name' attribute

    # Method to make the animal speak
    def speak(self):
        return f"{self.name} says hello!"  # Return a greeting with the 'name'

# Define a child class named 'Dog' that inherits from 'Animal'
class Dog(Animal):
    # Constructor for Dog class, takes 'name' and 'breed' parameters
    def __init__(self, name, breed):
        super().__init__(name)  # Call the constructor of the parent class ('Animal')
        self.breed = breed  # Assign the 'breed' parameter to the 'breed' attribute

# Create an instance of the 'Dog' class with the name "Charlie" and breed "Bulldog"
dog = Dog("Charlie", "Bulldog")

# Print the 'breed' attribute of the 'dog' object
print(dog.breed)  # Output will be "Bulldog"
```

### 4. Creating a property

```python
# Define a Circle class
class Circle:
    # Constructor that takes a radius as a parameter
    def __init__(self, radius):
        # Initialize the private attribute _radius using the radius property
        self.radius = radius

    # Define a property for getting the radius
    @property
    def radius(self):
        return self._radius

    # Define a setter for the radius property
    @radius.setter
    def radius(self, value):
        # Check if the provided value is non-negative
        if value >= 0:
            # If it's non-negative, set the private attribute _radius
            self._radius = value
        else:
            # If it's negative, raise a ValueError with an error message
            raise ValueError("Radius must be positive")

# Create an instance of the Circle class with a radius of 5
circle = Circle(5)

# Access and print the radius property (getter)
print(circle.radius)

# Set the radius property to 10 (invoking the setter)
circle.radius = 10

# Access and print the updated radius property
print(circle.radius)
```

### 5. Encapsulation  Private members

```python
class MyClass:
    def __init__(self):
        # Public variable - Accessible from anywhere
        self.public = "Public"

        # Protected variable - Convention to treat it as non-public, but accessible
        self._protected = "Protected"

        # Private variable - Name mangling is applied, not directly accessible
        self.__private = "Private"

obj = MyClass()

# Accessing the public variable, this works fine
print(obj.public)

# Accessing the protected variable (not recommended), it's accessible
print(obj._protected)

# Attempting to access the private variable (name mangling applied)
# This will result in an AttributeError, but there's a way to access it
# print(obj.__private)

# To access the private variable, you can use the name mangling technique:
print(obj._MyClass__private)
```
