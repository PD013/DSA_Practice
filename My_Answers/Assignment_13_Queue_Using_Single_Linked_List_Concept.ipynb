```python
class Node:
    def __init__(self,data=None,next=None):  #Here data = item
        self.data = data
        self.next = next
class Queue:
    def __init__(self):
        self.front = None
        self.rear = None
        self.counter = 0

    def is_empty(self):
        return self.front is None

    def enqueue(self,data):
        n = Node(data)
        self.counter +=1
        if self.is_empty():
            self.front = n
            self.rear = n
        else:
           self.rear.next = n
           self.rear = n

    def dequeue(self):
        if not self.is_empty():
            self.front = self.front.next
            self.counter -=1
        else:
            raise IndexError("The value Underflow")

    def get_front(self):
        return self.front.data

    def get_rear(self):
        return self.rear.data

    def size(self):
        return self.counter

q13 = Queue()
q13.enqueue(1)
print("This is the current front element",q13.get_front())
print("This is the current rear element",q13.get_rear())
print()
print()

q13.enqueue(2)
print("This is the current front element",q13.get_front())
print("This is the current rear element",q13.get_rear())
print()
print()

q13.enqueue(3)
print("This is the current front element",q13.get_front())
print("This is the current rear element",q13.get_rear())
print()
print()

q13.enqueue(4)
print("This is the current front element",q13.get_front())
print("This is the current rear element",q13.get_rear())
print()
print()

q13.dequeue()
print("This is the current front element",q13.get_front())
print("This is the current rear element",q13.get_rear())
print()
print()

print("This is the Size ",q13.size())

```
