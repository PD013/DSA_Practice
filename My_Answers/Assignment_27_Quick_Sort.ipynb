```python
# This is sir's Code
def quick_sort(list1):
    if len(list1)<=1: # Checking if more than one element in list
        return list1
    else:
        pivot=list1[0] # This is 1st element think as its sorted
        lesser=[x for x in list1[1:] if x<=pivot] # making a list where all small number than pivot is there
        greater=[x for x in list1[1:] if x>pivot] # making a list where all bigger number than pivot is there
        return quick_sort(lesser)+[pivot]+quick_sort(greater) # here using recursion and concationation of list

mylist=[53,11,72,68,41,25,18,37,44,80]
mylist=quick_sort(mylist)
print(mylist)
```
